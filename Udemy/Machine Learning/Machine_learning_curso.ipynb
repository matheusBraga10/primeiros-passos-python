##Classificação:
Se trata de prever um rótulo
- É de um tipo ou de outro?



##Regressão:
Se trata de prever um valor ou quantidade
- Quanto vou vender mês que vêm com base no meu histórico de vendas?


```python
# Exemplo Classificação
portatilA = [0,1,1]
portatilB = [1,1,1]
portatilC = [1,0,1]
portatilD = [1,0,0]
portatilE = [1,1,0]
portatilF = [0,1,0]
```

```python
portatilC
```

```python
database = [portatilA, portatilB, portatilC, portatilD, portatilE, portatilF]
database
```

```python
rotulo = ['Smartphone', 'Smartphone','Smartphone','Tablet','Tablet','Tablet']
rotulo
```

```python
from sklearn.dummy import DummyClassifier
# Biblioteca comum de machine learning

modelo = DummyClassifier()
modelo.fit(database, rotulo) # .fit é utilizado para aprender relacionar os dados que são passados (database, rotulo)
```

```python
desconhecido01 = [0,0,1]
desconhecido02 = [0,1,0]
desconhecido03 = [1,0,0]
desconhecido04 = [1,1,1]
desconhecido05 = [1,0,1]
desconhecido06 = [1,1,0]
desconhecido07 = [0,1,1]
desconhecido08 = [0,0,0]
```

```python
portateisDesconhecidos = [desconhecido01, desconhecido02, desconhecido03, desconhecido04, desconhecido05, desconhecido06, desconhecido07, desconhecido08]

portateisDesconhecidos
```

```python
y_pred = modelo.predict(portateisDesconhecidos) # Predição
```

```python
y_True = ['Smartphone','Tablet','Tablet','Smartphone','Smartphone','Tablet','Smartphone','Tablet']
y_True
```

```python
print('Resultados obtidos: ')
print(y_pred)
print('\nResultados esperados: ')
print(y_True)
```

```python
# verificando se o classificador está funcionando corretamente

from sklearn import metrics

print('Precisão: ' + str(metrics.precision_score(y_True, y_pred, average='macro')))
```
