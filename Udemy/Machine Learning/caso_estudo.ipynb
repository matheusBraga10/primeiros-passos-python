# Caso Estudo


## Início
- caso_estudo.xlsx

```python
import pandas as pd
```

```python
pessoas = pd.read_excel('caso_estudo.xlsx', sheet_name='pessoas')
covid = pd.read_excel('caso_estudo.xlsx', sheet_name='registros_covid')
```

```python
print(pessoas.head(10)) # imprime o cabeçalho do DataFrame (planilha)
print('\n')
print(pessoas.tail(5)) # imprime os ultimos dados da tabela
print('\n')
print(pessoas.sample(3)) # imprime dados aleatóriamente
print('\n')
print(pessoas.describe()) # imprime um resumo dos dados 
```

```python
# Criano filtros de dados

print(pessoas[pessoas.idade>40]) # Filtrando pessoas com idades acima de 40
print('\n')
print(pessoas[pessoas.idade>40].count()) # Apenas a quantidade
print('\n')
print(pessoas.nome[pessoas.idade==48]) # Filtrando por colunas (nome * numero)
```

```python
print(pessoas.idade[pessoas.sexo=='F'].mean()) # Média da idade de pessoas do sexo F
```

```python
print(pessoas[pessoas.nome.str[0]=='A']) # Filtrando pessoas que começam o nome pela letra A
print('\n')
print(pessoas[pessoas.nome.str[0:3]=='Ana'])
```

```python
# Combinando dados de um DataFrame

print(pessoas.groupby('sexo').idade.mean()) # groupby agrupa dados 
print('\n')
print(pessoas.groupby('sexo').idade.count())
```

```python
pessoas['inicial'] = pessoas.nome.str[0]
print(pessoas)
```

```python
pessoas.groupby('inicial').idade.mean()
```

## Join

```python
pessoas.head(2)
```

```python
covid.head(2)
```

```python
pessoas = pessoas.set_index('id')
covid = covid.set_index('id')
```

```python
df = pessoas.join(covid.set_index('id_pessoa'))
df
```

```python
df = df.fillna(0)
df
```

```python
df.groupby('severidade').idade.mean()
```

## Novas Features

```python
df['faixa_etaria'] = 'Jovem'
df
```

```python
df.loc[df.idade>24,'faixa_etaria'] = 'Adulto'
df.loc[df.idade>45,'faixa_etaria'] = 'Adulto2'
df.loc[df.idade>60,'faixa_etaria'] = 'Idoso'
df.loc[df.severidade=='baixa', 'severidade'] = 1
df.loc[df.severidade=='alta', 'severidade'] = 2
df['severidade'] = df['severidade'].astype('int')
df.groupby(['faixa_etaria']).idade.mean()
```
