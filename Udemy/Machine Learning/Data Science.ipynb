## Dicionários

```python
nomes = ['Rafael','Maria', 'Ana']  # Lista - modificável
idades = [10,20,15] # Lista - modificável

sexo = ('M', 'F', 'O') # Tupla - NÃO modificável (não poderá ser alterada ou adicionar valores)

cadastro = {'nomes': nomes,
            'idades': idades,
            'cidade': ['A','B','C']} # Dicionário - Indice modificável
```

```python
print(nomes[0])
print(idades[0])
print(sexo[0]) # Uso de colchetes para acessar a tupla
print(cadastro['nomes'][0])
print(cadastro['idades'])
print(sum(cadastro['idades'])) # Soma de idades
```

  ## For

  ## While

  ## If

```python
# for
for nome in nomes:
  print(nome)

print('\n')

for idade in idades:
  if idade % 2 == 0: # Se for par
    print(idade)

print('\n')

for i in range(0,5):
  print(i)
```

```python
# while
i = 0
idade = idades[i]
while idade <= 10:
  print(idade)
  i = i + 1
  idade = idades[i]

```

```python
# comandos inline

soma = 0
for i in range(0,100):
  if i%2 == 0:
    soma = soma + i
print(soma)

sum([i if i%2==0 else 0 for i in range(0,100)]) # Mesmo código, em uma unica linha
```

## Funções

```python
# import minha_funcao - caso tenha essa funcao em outro arquivo chamado minha_funcao.py (só funciona fora do colab)

def faixa_idade(idade):
  if idade <= 12:
    return 'Criança'
  elif idade < 18:
    return 'Adolescente'
  elif idade <= 80:
    return 'Adulto'
  else:
    return 'Idoso'

def soma_quadrados(n1,n2):
  return n1**2 + n2**2


faixa_idade(int(input(('Digite a Idade: '))))
soma_quadrados(2,4)
```
